## Projet Pi 
Quelle est la bonne valeur de Pi que je peux calculer?
* Méthode de Wallis : 
$$ \frac{\pi}{2} = \prod_{k = 1}^\infty  \frac{(2 k)^2}{(2 k )^2 -1}$$
* Méthode de Leibiniz : 
$$\pi = 4 \sum_{k = 0}^\infty \frac{(-1)^k}{2 k - 1} $$

```python
S_Leibniz = 0
for k in range(0, 1001):
    S_Leibniz+= ((-1)**k)/(2 * k + 1)
pi_Leibniz = S_Leibniz*4
print(pi_Leibniz)
err_leibniz = abs(pi_math - pi_Leibniz)
print(err_leibniz)
```

```python
P_Wallis = 1
for k in range(1, 1001):
    P_Wallis *= (2*k)**2/((2*k)**2 -1)
pi_Wallis = 2 * P_Wallis
print(pi_Wallis)
err_Wallis = abs(pi_math-pi_Wallis)
print(err_Wallis)
```

```python
def Pi_Leibniz(N):
    S_Leibniz = 0
    for k in range(0, N+1):
        S_Leibniz+= ((-1)**k)/(2 * k + 1)
    pi_Leibniz = S_Leibniz*4
    return pi_Leibniz
```

```python
def Pi_Wallis(N):
    P_Wallis = 1
    for k in range(1, N+1):
        P_Wallis *= (2*k)**2/((2*k)**2 -1)
    pi_Wallis = 2 * P_Wallis
    return(pi_Wallis)
```

```python
Pi_Wallis(1000)
```

```python
Pi_Leibniz(1000)
```

```python
L_Leibniz = [Pi_Leibniz(n) for n in range(2, 501)]
L_n = list(range(2, 501))
print(L_Leibniz)
print(len(L_Leibniz)== len(L_n))
```

```python
L_Wallis = [Pi_Wallis(n) for n in range(2, 501)]
```

```python
import matplotlib.pyplot as plt
plt.plot(L_n, L_Leibniz, label = "Leibniz")
plt.plot(L_n, L_Wallis, label = "Wallis")
plt.legend()
plt.xlabel("n")
plt.ylabel("f(n)")
plt.savefig("pi.png")
```
